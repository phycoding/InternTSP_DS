```python
#To analysing total no of order from different states
state = df.value_counts()
plt.figure(figsize=(16,16))
sns.countplot(x=df["State"])
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(5,5))
df.groupby('Category')['Profit'].sum().plot.pie(shadow=True)
plt.title('Categrory Wise Profit',Fontsize=21)
```

```python
#to see region wise different sub_categories order
plt.figure(figsize=(15,5))
sns.countplot(x='Sub-Category',hue="Region",data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
#Statewise Sales analysis
plt.figure(figsize=(20,20))
df.groupby("State")['Sales'].sum().plot.pie()
plt.title("State wise Sales analysis",fontsize=29)
```

```python
pd.crosstab(df["Region"],df["Category"]).plot(kind='bar')
```

```python
#profit vs Discount
sns.lineplot(x=df["Discount"],y=df["Profit"])
```

```python
#Profit vs Quantity
sns.lineplot(y=df["Profit"],x=df["Quantity"])
```

```python
#Discounts vs sales
sns.lineplot(y=df['Sales'],x=df['Discount'])
```
